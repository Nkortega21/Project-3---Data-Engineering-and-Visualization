```python
import pandas as pd
```

```python
df_raw = pd.read_csv("Resources/amazon_sale_report.csv",low_memory = False);
```

```python
df_raw.head()
```

```python
df_raw.columns
```

```python
df_raw.info()
```

```python
df_raw['ship-postal-code']= df_raw['ship-postal-code'].astype(str)
df_raw.info()
```

```python
df_raw['ship-postal-code'] = df_raw['ship-postal-code'].str.replace(r'\.0$', '', regex=True)
df_raw.head()
```

```python
orders_df = df_raw[['Order ID', 'Date', 'Status', 'Fulfilment', 'Sales Channel ', 'ship-service-level', 'B2B']].copy()
orders_df.columns = ['OrderID', 'Date', 'Status', 'Fulfilment', 'SalesChannel', 'ShipServiceLevel', 'B2B']

# Ensure unique OrderID
orders_df.drop_duplicates(subset='OrderID', inplace=True)
orders_df.head()
```

```python
# Products table
products_df = df_raw[['Order ID','SKU', 'Style', 'Category', 'Amount', 'currency']].copy()
products_df.columns = ['OrderID','SKU', 'Style', 'Category', 'Amount', 'Currency']

# Ensure unique SKU
products_df.drop_duplicates(subset='SKU', inplace=True)
products_df.head()
```

```python
# ShippingDetails table
shipping_df = df_raw[['Order ID', 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country']].copy()
shipping_df.columns = ['OrderID', 'ShipCity', 'ShipState', 'ShipPostalCode', 'ShipCountry']

shipping_df.head()
```

```python
# Making dataframes into our csv's for import into SQL
orders_df.to_csv("ETL_csv/orders_table.csv",index = False)
products_df.to_csv("ETL_csv/products_table.csv",index = False)
shipping_df.to_csv("ETL_csv/shipping_table.csv",index = False)
```
