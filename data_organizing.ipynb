```python
import pandas as pd
import numpy as np 
```

```python
# Load the Amazon sales report CSV file into a DataFrame
df_raw = pd.read_csv("Resources/amazon_sale_report.csv",low_memory = False);
```

```python
# Remove all duplicate rows from the DataFrame, keeping only unique rows
df_cleaned = df_raw.drop_duplicates(keep=False)

# Display the first few rows of the original DataFrame to check the data
df_raw.head()
```

```python
# Display the names of all columns in the DataFrame
df_raw.columns
```

```python
# Display information about the DataFrame
df_raw.info()
```

```python
# Identify duplicate rows in the DataFrame
duplicates = df_raw[df_raw.duplicated(keep=False)]

# Print all duplicate rows
if not duplicates.empty:  # Check if the duplicates DataFrame is not empty
    print("Duplicate rows based on all columns:")
    print(duplicates)  # Print the duplicate rows
else:
    print("No duplicates found across all columns.")  # Inform if no duplicates exist
```

```python
# Convert the 'ship-postal-code' column to string data type
df_raw['ship-postal-code'] = df_raw['ship-postal-code'].astype(str)

# Display information about the DataFrame, including the data types of each column
df_raw.info()
```

### above variables whether we should add sth like 0.00 etc.

```python
# Remove any trailing '.0' from the 'ship-postal-code' column values
df_raw['ship-postal-code'] = df_raw['ship-postal-code'].str.replace(r'\.0$', '', regex=True)

# Strip whitespace from the column names
df_raw.columns = df_raw.columns.str.strip()

# Display the first few rows of the cleaned DataFrame
df_raw.head()
```

```python
# Create a new DataFrame by selecting specific columns and removing duplicates based on 'Order ID'
orders_df = df_raw[['Order ID', 'Date', 'Status', 'Fulfilment', 'Sales Channel', 'ship-service-level', 'B2B']].drop_duplicates(subset=['Order ID'])

# Rename the columns for better readability and consistency
orders_df.columns = ['OrderID', 'Date', 'Status', 'Fulfilment', 'SalesChannel', 'ShipServiceLevel', 'B2B']

# Display the first few rows of the new DataFrame
orders_df.head()
```

```python
# Create a new DataFrame by selecting specific columns and removing duplicates based on 'Order ID' and 'SKU'
products_df = df_raw[['SKU', 'Order ID', 'Style', 'Category', 'Qty', 'currency', 'Amount']].drop_duplicates(subset=['Order ID', 'SKU'])

# Rename the columns for better readability and consistency
products_df.columns = ['SKU', 'OrderID', 'Style', 'Category', 'Qty', 'Currency', 'Amount']

# Display the first few rows of the new DataFrame
products_df.head()
```

```python
# Create a new DataFrame by selecting specific columns and removing duplicates based on 'Order ID'
shipping_df = df_raw[['Order ID', 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country']].drop_duplicates(subset=['Order ID'])

# Rename the columns for better readability and consistency
shipping_df.columns = ['OrderID', 'ShipCity', 'ShipState', 'ShipPostalCode', 'ShipCountry']

# Display the first few rows of the new DataFrame
shipping_df.head()
```

```python
# Making dataframes into our csv's for import into SQL
orders_df.to_csv("ETL_csv/orders_table.csv",index = False)
products_df.to_csv("ETL_csv/products_table.csv",index = False)
shipping_df.to_csv("ETL_csv/shipping_table.csv",index = False)
```
