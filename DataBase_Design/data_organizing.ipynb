### Adding Latitude/Longitude with Shipping State to create a Leaflet Visual

```python
def get_lat_long_google(state_name, country="India", api_key=None):
    # Construct the query as State + Country to ensure you get the correct location
    location_query = f"{state_name}, {country}"
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={location_query}&key={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if data['results']:
            location = data['results'][0]['geometry']['location']
            return location['lat'], location['lng']
        else:
            return None, None
    else:
        return None, None
```
