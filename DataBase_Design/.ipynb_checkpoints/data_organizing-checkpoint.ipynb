```python
# Convert the 'ship-postal-code' column to string data type
df_raw['ship-postal-code'] = df_raw['ship-postal-code'].astype(str)

# Display information about the DataFrame, including the data types of each column
df_raw.info()
```

### above variables whether we should add sth like 0.00 etc.

```python
# Remove any trailing '.0' from the 'ship-postal-code' column values
df_raw['ship-postal-code'] = df_raw['ship-postal-code'].str.replace(r'\.0$', '', regex=True)

# Strip whitespace from the column names
df_raw.columns = df_raw.columns.str.strip()

# Display the first few rows of the cleaned DataFrame
df_raw.head()
```

```python
# Create a new DataFrame by selecting specific columns and removing duplicates based on 'Order ID'
orders_df = df_raw[['Order ID', 'Date', 'Status', 'Fulfilment', 'Sales Channel', 'ship-service-level', 'B2B']].drop_duplicates(subset=['Order ID'])

# Rename the columns for better readability and consistency
orders_df.columns = ['OrderID', 'Date', 'Status', 'Fulfilment', 'SalesChannel', 'ShipServiceLevel', 'B2B']

# Display the first few rows of the new DataFrame
orders_df.head()
```

```python
# Create a new DataFrame by selecting specific columns and removing duplicates based on 'Order ID' and 'SKU'
products_df = df_raw[['SKU', 'Order ID', 'Style', 'Category', 'Qty', 'currency', 'Amount']].drop_duplicates(subset=['Order ID', 'SKU'])

# Rename the columns for better readability and consistency
products_df.columns = ['SKU', 'OrderID', 'Style', 'Category', 'Qty', 'Currency', 'Amount']

# Display the first few rows of the new DataFrame
products_df.head()
```

```python
# Create a new DataFrame by selecting specific columns and removing duplicates based on 'Order ID'
shipping_df = df_raw[['Order ID', 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country']].copy()

# Rename the columns for better readability and consistency
shipping_df.columns = ['OrderID', 'ShipCity', 'ShipState', 'ShipPostalCode', 'ShipCountry']
shipping_df['ShipCity'] = shipping_df['ShipCity'].str.title()
shipping_df['ShipState'] = shipping_df['ShipState'].str.title()


# Display the first few rows of the new DataFrame
shipping_df.head()
```

```python
# Step 2: Cleaned list of Indian states and union territories
cleaned_states_ut = [
    'Maharashtra', 'Karnataka', 'Puducherry', 'Tamil Nadu', 'Uttar Pradesh',
    'Chandigarh', 'Telangana', 'Andhra Pradesh', 'Rajasthan', 'Delhi',
    'Haryana', 'Assam', 'Jharkhand', 'Chhattisgarh', 'Odisha', 'Kerala',
    'Madhya Pradesh', 'West Bengal', 'Nagaland', 'Gujarat', 'Uttarakhand',
    'Bihar', 'Jammu & Kashmir', 'Punjab', 'Himachal Pradesh', 'Arunachal Pradesh',
    'Manipur', 'Goa', 'Meghalaya', 'Tripura', 'Ladakh', 'Dadra and Nagar Haveli and Daman and Diu',
    'Sikkim', 'Andaman & Nicobar Islands', 'Lakshadweep'
]

# Step 3: Define a mapping of common errors or abbreviations to correct names
state_corrections = {
    'Pondicherry': 'Puducherry',
    'Orissa': 'Odisha',
    'Rajsthan': 'Rajasthan',
    'Rajshthan': 'Rajasthan',
    'Rj': 'Rajasthan',
    'Pb': 'Punjab',
    'Nl': 'Nagaland',
    'New Delhi': 'Delhi',
    'Punjab/Mohali/Zirakpur': 'Punjab',
    'Apo': None,  # Drop these rows if it's not valid
    'Ar': 'Arunachal Pradesh',
    'Dadra And Nagar': 'Dadra and Nagar Haveli and Daman and Diu',
    # Handle missing or NaN
    None: None,  # Retain NaNs as is or drop them later
    'nan': None  # Handle missing values
}

# Step 4: Update the 'shipstate' column
# Apply the corrections map and filter by cleaned_states_ut list
cleaned_shipping_df['ShipState'] = cleaned_shipping_df['ShipState'].replace(state_corrections)

# Step 5: Ensure that only valid states/UTs are retained
# If any values in 'shipstate' are not in the cleaned_states_ut list, set them to NaN
cleaned_shipping_df['ShipState'] = cleaned_shipping_df['ShipState'].apply(lambda x: x if x in cleaned_states_ut else None)

# Step 6: Optionally, drop rows with missing or invalid states
# df = df.dropna(subset=['shipstate'])

# Step 7: Save the updated dataset
cleaned_shipping_df.head()
```

### Adding Latitude/Longitude with Shipping State to create a Leaflet Visual

```python
def get_lat_long_google(state_name, country="India", api_key=None):
    # Construct the query as State + Country to ensure you get the correct location
    location_query = f"{state_name}, {country}"
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={location_query}&key={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if data['results']:
            location = data['results'][0]['geometry']['location']
            return location['lat'], location['lng']
        else:
            return None, None
    else:
        return None, None
```

```python
# Convert the dictionary into a DataFrame
states_info_df = pd.DataFrame.from_dict(state_coordinates, orient='index', columns=['Latitude', 'Longitude']).reset_index()

# Rename the index column to "State"
states_info_df = states_info_df.rename(columns={'index': 'ShipState'})

# Display the DataFrame
states_info_df
```
